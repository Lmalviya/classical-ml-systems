# Math Prerequisites → Linear Algebra | Chapter 0: Vectors & Projections

> **Why this chapter exists:** PCA is fundamentally an operation of *projecting* data onto new directions. Before we can understand what PCA does, we need to understand what a vector is, how to measure angles and distances between vectors, and most importantly — what it means to *project* one vector onto another. Every calculation in PCA traces back to the projection formula derived in this chapter.

---

## 1. Vectors: Definition, Notation & Operations

### **What is a Vector?**
A vector is an ordered list of numbers that represents a point or a direction in space.

$$\vec{x} = \begin{bmatrix} x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix} \in \mathbb{R}^n$$

Two interpretations — both valid and useful:
- **Point:** $\vec{x}$ is a location in $n$-dimensional space (a data sample with $n$ features).
- **Direction:** $\vec{x}$ is an arrow from the origin pointing toward $(x_1, x_2, ..., x_n)$.

In Machine Learning, a data point with $d$ features is a vector $\vec{x} \in \mathbb{R}^d$. A dataset of $N$ such points is a matrix $X \in \mathbb{R}^{N \times d}$ (each row is a data point).

### **Basic Operations**

**Addition:** Component-wise. Geometrically: tip-to-tail rule.
$$\vec{u} + \vec{v} = \begin{bmatrix} u_1 + v_1 \\ u_2 + v_2 \\ \vdots \end{bmatrix}$$

**Scalar Multiplication:** Stretch or shrink the vector. Negative scalar = flip direction.
$$c \cdot \vec{u} = \begin{bmatrix} c u_1 \\ c u_2 \\ \vdots \end{bmatrix}$$

### **Vector Norms: Measuring Length**
The **norm** of a vector measures its length (distance from origin).

| Norm | Formula | Intuition | Used For |
| :--- | :--- | :--- | :--- |
| $L_1$ norm | $\|\vec{x}\|_1 = \sum_i |x_i|$ | Manhattan distance (city blocks) | Sparse solutions (Lasso) |
| $L_2$ norm | $\|\vec{x}\|_2 = \sqrt{\sum_i x_i^2}$ | Euclidean distance (straight line) | PCA, most geometry |
| $L_\infty$ norm | $\|\vec{x}\|_\infty = \max_i |x_i|$ | Chebyshev distance (chessboard) | Robustness |

**The $L_2$ norm is the standard** in PCA and most of linear algebra. When we say "length" of a vector, we mean $\|\vec{x}\|_2$.

$$\|\vec{x}\|_2 = \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2}$$

### **Unit Vectors**
A **unit vector** has length exactly 1:
$$\hat{u} = \frac{\vec{u}}{\|\vec{u}\|_2}$$

Unit vectors represent **pure directions** — no magnitude information, only direction. In PCA, the principal components are always unit vectors (they define directions of maximum variance).

## 2. The Dot Product: Measuring Alignment

### **Algebraic Definition**
The dot product (inner product) of two vectors $\vec{u}, \vec{v} \in \mathbb{R}^n$:

$$\vec{u} \cdot \vec{v} = \sum_{i=1}^n u_i v_i = u_1 v_1 + u_2 v_2 + \cdots + u_n v_n$$

In matrix notation, if $\vec{u}$ and $\vec{v}$ are column vectors:
$$\vec{u} \cdot \vec{v} = \vec{u}^T \vec{v}$$

### **Geometric Definition**
The dot product has a beautiful geometric interpretation:

$$\vec{u} \cdot \vec{v} = \|\vec{u}\| \cdot \|\vec{v}\| \cdot \cos\theta$$

where $\theta$ is the angle between $\vec{u}$ and $\vec{v}$.

Rearranging: $\cos\theta = \frac{\vec{u} \cdot \vec{v}}{\|\vec{u}\| \cdot \|\vec{v}\|}$

**What $\cos\theta$ tells you:**

| $\cos\theta$ | $\theta$ | Interpretation |
| :---: | :---: | :--- |
| $= 1$ | $0°$ | Vectors point in the exact same direction |
| $> 0$ | $< 90°$ | Vectors point in roughly the same direction |
| $= 0$ | $90°$ | Vectors are **perpendicular (orthogonal)** |
| $< 0$ | $> 90°$ | Vectors point in roughly opposite directions |
| $= -1$ | $180°$ | Vectors point in exactly opposite directions |

### **The Dot Product as a Projection Signal**
This is the key insight that drives PCA:

$$\vec{u} \cdot \hat{v} = \|\vec{u}\| \cos\theta$$

When $\vec{v}$ is a unit vector $\hat{v}$: the dot product $\vec{u} \cdot \hat{v}$ is exactly the **signed length of the shadow** of $\vec{u}$ onto the direction $\hat{v}$. This is the definition of projection — covered in detail in Section 4.

### **Properties of the Dot Product**
- **Commutative:** $\vec{u} \cdot \vec{v} = \vec{v} \cdot \vec{u}$
- **Distributive:** $\vec{u} \cdot (\vec{v} + \vec{w}) = \vec{u} \cdot \vec{v} + \vec{u} \cdot \vec{w}$
- **Self-dot = squared norm:** $\vec{u} \cdot \vec{u} = \|\vec{u}\|^2$

## 3. Orthogonality: The Foundation of PCA's New Axes

### **Definition**
Two vectors $\vec{u}$ and $\vec{v}$ are **orthogonal** if their dot product is zero:
$$\vec{u} \perp \vec{v} \iff \vec{u} \cdot \vec{v} = 0 \iff \theta = 90°$$

Orthogonal vectors carry **no shared information** — they are completely independent directions.

### **Why Orthogonality Matters in PCA**
PCA finds a sequence of directions (principal components) $\vec{w}_1, \vec{w}_2, ..., \vec{w}_k$ such that:
- Each $\vec{w}_i$ is a unit vector: $\|\vec{w}_i\| = 1$
- Every pair is orthogonal: $\vec{w}_i \cdot \vec{w}_j = 0$ for $i \neq j$
- Together they form an **orthonormal set**

Why do we need orthogonality? Because orthogonal PCs are **uncorrelated** — each one captures variance that the previous PCs haven't already captured. If they weren't orthogonal, two PCs could contain overlapping information.

### **Orthonormal Basis**
A set of vectors $\{\vec{e}_1, \vec{e}_2, ..., \vec{e}_n\}$ is an **orthonormal basis** if:
1. Each vector is a unit vector: $\|\vec{e}_i\| = 1$
2. All pairs are orthogonal: $\vec{e}_i \cdot \vec{e}_j = 0$ for $i \neq j$
3. Together they span $\mathbb{R}^n$ — every vector in $\mathbb{R}^n$ can be written as a linear combination of them

The standard basis $\{\vec{e}_1 = [1,0,0...], \vec{e}_2 = [0,1,0...], ...\}$ is the most familiar orthonormal basis. PCA finds a **new** orthonormal basis that is better aligned with the data's variance.

**Elegant property:** For orthonormal basis $Q = [\vec{q}_1 | \vec{q}_2 | ... | \vec{q}_n]$:
$$Q^T Q = I \quad \text{and} \quad Q Q^T = I$$
i.e., $Q^{-1} = Q^T$. This is an extremely useful property we will use repeatedly.

## 4. Projection: The Core Operation Behind PCA

### **What is Projection?**
Projecting vector $\vec{u}$ onto vector $\vec{v}$ means: *find the component of $\vec{u}$ that lies along the direction of $\vec{v}$*. Geometrically: drop a perpendicular from the tip of $\vec{u}$ onto the line defined by $\vec{v}$.

```
          ┐ ← vec{u}
         /|
        / |  ← the perpendicular (error)
       /  |
──────/───┘─────────────→ direction of vec{v}
    origin  proj
```

### **Deriving the Projection Formula**
We want to find the scalar $\alpha$ such that $\alpha \vec{v}$ is the projection of $\vec{u}$ onto $\vec{v}$.

**Key condition:** The error vector $\vec{u} - \alpha\vec{v}$ must be perpendicular to $\vec{v}$:
$$(\vec{u} - \alpha\vec{v}) \cdot \vec{v} = 0$$
$$\vec{u} \cdot \vec{v} - \alpha \vec{v} \cdot \vec{v} = 0$$
$$\alpha = \frac{\vec{u} \cdot \vec{v}}{\vec{v} \cdot \vec{v}} = \frac{\vec{u} \cdot \vec{v}}{\|\vec{v}\|^2}$$

The **projection vector** (the shadow of $\vec{u}$ along $\vec{v}$):
$$\text{proj}_{\vec{v}} \vec{u} = \alpha \vec{v} = \frac{\vec{u} \cdot \vec{v}}{\|\vec{v}\|^2} \vec{v}$$

The **scalar projection** (the signed length of the shadow):
$$\text{scalar proj} = \alpha \|\vec{v}\| = \frac{\vec{u} \cdot \vec{v}}{\|\vec{v}\|}$$

### **When $\vec{v}$ is a Unit Vector**
If $\hat{v}$ is a unit vector: $\|\hat{v}\| = 1$, so $\|\hat{v}\|^2 = 1$.
$$\text{proj}_{\hat{v}} \vec{u} = (\vec{u} \cdot \hat{v}) \hat{v}$$
$$\text{scalar proj} = \vec{u} \cdot \hat{v}$$

This is beautifully simple: **to project onto a unit vector, just take the dot product**. This is why PCA uses unit-length principal components — projecting each data point onto a PC is just a dot product.

### **Projection Matrix**
The projection of $\vec{u}$ onto $\hat{v}$ can also be written as a **matrix operation**:
$$\text{proj}_{\hat{v}} \vec{u} = (\hat{v}\hat{v}^T) \vec{u} = P\vec{u}$$

where $P = \hat{v}\hat{v}^T \in \mathbb{R}^{n \times n}$ is the **projection matrix** onto the subspace spanned by $\hat{v}$.

**Properties of projection matrix $P$:**
- **Idempotent:** $P^2 = P$ (projecting twice = projecting once)
- **Symmetric:** $P^T = P$
- **Rank 1:** Projects everything onto a single line

For projection onto a $k$-dimensional subspace spanned by orthonormal vectors $V_k = [\hat{v}_1 | ... | \hat{v}_k]$:
$$P_k = V_k V_k^T$$

This is exactly the reconstruction formula in PCA: $\hat{X} = X V_k V_k^T$.

### **The Residual (Error)**
The part of $\vec{u}$ **not** captured by the projection:
$$\vec{r} = \vec{u} - \text{proj}_{\hat{v}} \vec{u} = \vec{u} - (\vec{u} \cdot \hat{v})\hat{v} = (I - P)\vec{u}$$

The residual $\vec{r}$ is always orthogonal to $\hat{v}$: $\vec{r} \cdot \hat{v} = 0$.

**This is the reconstruction error in PCA** — when we project data onto $k$ PCs, the residual is the information lost. PCA minimizes this residual.

### **Connection to PCA (Preview)**
In PCA:
- Data matrix $X \in \mathbb{R}^{N \times d}$ — each row = one data point $\vec{x}_i \in \mathbb{R}^d$
- Principal component direction: $\hat{w} \in \mathbb{R}^d$ (unit vector)
- Score of sample $i$ on this PC: $z_i = \vec{x}_i \cdot \hat{w}$ (a dot product = scalar projection)
- Reconstruction of sample $i$: $\hat{x}_i = z_i \hat{w} = (\vec{x}_i \cdot \hat{w})\hat{w}$ (projection vector)
- Reconstruction error for sample $i$: $\|\vec{x}_i - \hat{x}_i\|^2$

PCA finds $\hat{w}$ that minimizes the total reconstruction error $\sum_i \|\vec{x}_i - \hat{x}_i\|^2$.

## 5. ✍️ Worked Examples

### **Example 1: Dot Product and Angle**
Given $\vec{u} = [3, 4]$ and $\vec{v} = [1, 0]$:

- Dot product: $\vec{u} \cdot \vec{v} = 3 \times 1 + 4 \times 0 = 3$
- $\|\vec{u}\| = \sqrt{3^2 + 4^2} = \sqrt{25} = 5$, $\|\vec{v}\| = 1$
- $\cos\theta = \frac{3}{5 \times 1} = 0.6 \Rightarrow \theta = \arccos(0.6) \approx 53.1°$

### **Example 2: Scalar & Vector Projection**
Project $\vec{u} = [3, 4]$ onto $\vec{v} = [1, 0]$:

- $\vec{v}$ is already a unit vector: $\hat{v} = [1, 0]$
- Scalar projection: $\vec{u} \cdot \hat{v} = 3$
- Vector projection: $3 \times [1, 0] = [3, 0]$
- Residual: $[3, 4] - [3, 0] = [0, 4]$
- Check orthogonality: $[0, 4] \cdot [1, 0] = 0$ ✅

Interpretation: $\vec{u} = [3, 4]$ has a component of length 3 along the $x$-axis, and a component of length 4 along the $y$-axis (which is perpendicular to $x$).

### **Example 3: Projection onto a Non-Axis Direction**
Project $\vec{u} = [3, 4]$ onto $\vec{v} = [1, 1]$:

- Normalize: $\hat{v} = \frac{1}{\sqrt{2}}[1, 1]$
- Scalar projection: $\vec{u} \cdot \hat{v} = \frac{1}{\sqrt{2}}(3 + 4) = \frac{7}{\sqrt{2}} \approx 4.95$
- Vector projection: $\frac{7}{\sqrt{2}} \times \frac{1}{\sqrt{2}}[1, 1] = \frac{7}{2}[1, 1] = [3.5, 3.5]$
- Residual: $[3, 4] - [3.5, 3.5] = [-0.5, 0.5]$
- Check: $[-0.5, 0.5] \cdot [1, 1] = -0.5 + 0.5 = 0$ ✅

Note: The scalar projection onto the diagonal direction $[1,1]$ is 4.95 — *larger* than the projection onto either axis (3 or 4). The diagonal direction captures more of the vector's length. **This is exactly what PCA's first principal component does** — it finds the direction that maximizes the scalar projections (variance) of the data.

### **Example 4: 3D Projection onto 2D Plane**
Data point $\vec{x} = [2, 3, 5]$. PCA found two unit PCs:
$\hat{w}_1 = \frac{1}{\sqrt{2}}[1, 1, 0]$, $\hat{w}_2 = \frac{1}{\sqrt{2}}[1, -1, 0]$

- Score on PC1: $z_1 = \vec{x} \cdot \hat{w}_1 = \frac{1}{\sqrt{2}}(2 + 3 + 0) = \frac{5}{\sqrt{2}} \approx 3.54$
- Score on PC2: $z_2 = \vec{x} \cdot \hat{w}_2 = \frac{1}{\sqrt{2}}(2 - 3 + 0) = \frac{-1}{\sqrt{2}} \approx -0.71$
- Reconstruction: $\hat{x} = z_1 \hat{w}_1 + z_2 \hat{w}_2 = \frac{5}{2}[1,1,0] + \frac{-1}{2}[1,-1,0] = [2, 3, 0]$
- Residual: $[2,3,5] - [2,3,0] = [0, 0, 5]$ — the $z$-component is lost

## 6. Key Formulas Summary

| Formula | Expression | Used In |
| :--- | :--- | :--- |
| Dot product | $\vec{u} \cdot \vec{v} = \sum_i u_i v_i = \vec{u}^T \vec{v}$ | Inner product everywhere |
| Geometric dot | $\vec{u} \cdot \vec{v} = \|\vec{u}\|\|\vec{v}\|\cos\theta$ | Angle between vectors |
| $L_2$ norm | $\|\vec{u}\| = \sqrt{\vec{u} \cdot \vec{u}}$ | Distances, lengths |
| Unit vector | $\hat{u} = \vec{u}/\|\vec{u}\|$ | PC directions |
| Orthogonality | $\vec{u} \perp \vec{v} \iff \vec{u} \cdot \vec{v} = 0$ | PC independence |
| Scalar projection | $\vec{u} \cdot \hat{v}$ | PC scores |
| Vector projection | $(\vec{u} \cdot \hat{v})\hat{v}$ | Reconstructed data point |
| Projection matrix | $P = \hat{v}\hat{v}^T$; $P_k = V_k V_k^T$ | Reconstruction formula |
| Residual | $\vec{u} - (\vec{u}\cdot\hat{v})\hat{v} = (I-P)\vec{u}$ | Reconstruction error |

## 7. Implementation Playground

````python
# ─── CELL 1: Vector Operations & Norms ───────────────────────────────────────
import numpy as np

u = np.array([3.0, 4.0])
v = np.array([1.0, 0.0])

# Dot product - two equivalent ways
dot_manual = np.sum(u * v)
dot_matrix = u @ v  # u^T v notation
print(f"Dot product: {dot_manual} (manual) = {dot_matrix} (matrix)")

# Norms
l1 = np.sum(np.abs(u))
l2 = np.sqrt(np.sum(u**2))   # same as np.linalg.norm(u)
linf = np.max(np.abs(u))
print(f"L1={l1}, L2={l2}, Linf={linf}")

# Angle between vectors
cos_theta = (u @ v) / (np.linalg.norm(u) * np.linalg.norm(v))
theta_deg = np.degrees(np.arccos(cos_theta))
print(f"Angle between u and v: {theta_deg:.2f}°")

# Unit vector
u_hat = u / np.linalg.norm(u)
print(f"Unit vector of u: {u_hat}, length = {np.linalg.norm(u_hat):.6f}")
````

````python
# ─── CELL 2: Projection Formula — Manual & Visual ────────────────────────────
import numpy as np
import matplotlib.pyplot as plt

def project(u, v):
    """Project vector u onto vector v. Returns vector projection."""
    v_hat = v / np.linalg.norm(v)
    scalar = u @ v_hat          # scalar projection
    proj_vec = scalar * v_hat   # vector projection
    residual = u - proj_vec     # error / residual
    return proj_vec, residual, scalar

u = np.array([3.0, 4.0])
v = np.array([1.0, 1.0])  # diagonal direction

proj_vec, residual, scalar = project(u, v)
print(f"Scalar projection onto v: {scalar:.4f}")
print(f"Vector projection: {proj_vec}")
print(f"Residual: {residual}")
print(f"Orthogonality check (should be ~0): {residual @ (v/np.linalg.norm(v)):.10f}")

# Verify: ||u||^2 = ||proj||^2 + ||residual||^2 (Pythagorean theorem)
print(f"Pythagoras check: {np.linalg.norm(u)**2:.4f} = {np.linalg.norm(proj_vec)**2 + np.linalg.norm(residual)**2:.4f}")

# Visualization
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
origin = np.zeros(2)

ax.quiver(*origin, *u, color='blue', angles='xy', scale_units='xy', scale=1, label='u = [3,4]')
ax.quiver(*origin, *v/np.linalg.norm(v)*5, color='gray', angles='xy', scale_units='xy', scale=1, alpha=0.4, label='direction v')
ax.quiver(*origin, *proj_vec, color='green', angles='xy', scale_units='xy', scale=1, label=f'projection = {proj_vec.round(2)}')
ax.quiver(*proj_vec, *residual, color='red', angles='xy', scale_units='xy', scale=1, label=f'residual = {residual.round(2)}')

ax.set_xlim(-1, 5); ax.set_ylim(-1, 5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.axhline(0, color='k', lw=0.5); ax.axvline(0, color='k', lw=0.5)
ax.legend(fontsize=11); ax.set_title('Vector Projection: u onto v', fontsize=14)
plt.tight_layout(); plt.show()
````

````python
# ─── CELL 3: Projection Matrix — Properties & Subspace Projection ─────────────
import numpy as np

# Single unit vector → rank-1 projection matrix
v = np.array([1.0, 1.0])
v_hat = v / np.linalg.norm(v)

P = np.outer(v_hat, v_hat)  # v_hat @ v_hat.T
print("Projection matrix P:")
print(P)

# Property 1: Idempotent — P^2 = P
print(f"\nP^2 = P? {np.allclose(P @ P, P)}")

# Property 2: Symmetric — P^T = P
print(f"P^T = P? {np.allclose(P.T, P)}")

# Property 3: Rank 1
print(f"Rank of P: {np.linalg.matrix_rank(P)}")

# Apply to u — should give same as manual formula
u = np.array([3.0, 4.0])
proj_via_matrix = P @ u
proj_via_formula = (u @ v_hat) * v_hat
print(f"\nProjection via matrix: {proj_via_matrix}")
print(f"Projection via formula: {proj_via_formula}")
print(f"Same result? {np.allclose(proj_via_matrix, proj_via_formula)}")

# 2D subspace projection (two orthonormal vectors)
w1 = np.array([1., 0., 0.])  # x-axis
w2 = np.array([0., 1., 0.])  # y-axis
V_k = np.column_stack([w1, w2])  # 3x2 matrix

P_k = V_k @ V_k.T  # projects onto xy-plane
x = np.array([2., 3., 5.])
print(f"\n3D → 2D projection of {x}: {P_k @ x}")
print(f"(z-component dropped: {(P_k @ x)[2]:.1f})")
````

````python
# ─── CELL 4: Which Direction Maximizes Scalar Projections? (PCA Intuition) ────
import numpy as np
import matplotlib.pyplot as plt

# Create correlated 2D data — variance is larger along the diagonal
np.random.seed(42)
N = 200
# Data is elongated along the direction [1, 2]
t = np.random.randn(N)
noise = np.random.randn(N, 2) * 0.3
X = np.outer(t, [1, 2]) + noise   # shape (N, 2)
X = X - X.mean(axis=0)  # center

# Test different projection directions (angles from 0° to 180°)
angles = np.linspace(0, np.pi, 181)
variances = []
for a in angles:
    direction = np.array([np.cos(a), np.sin(a)])
    scores = X @ direction  # scalar projections for all N points
    variances.append(np.var(scores))

best_angle_idx = np.argmax(variances)
best_angle_deg = np.degrees(angles[best_angle_idx])
best_direction = np.array([np.cos(angles[best_angle_idx]), np.sin(angles[best_angle_idx])])

print(f"Direction with maximum variance: {best_angle_deg:.1f}° from x-axis")
print(f"Best direction vector: {best_direction.round(3)}")
print(f"True direction was [1,2], normalized: {np.array([1,2])/np.linalg.norm([1,2]):.3f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: variance vs angle
axes[0].plot(np.degrees(angles), variances, color='royalblue', lw=2)
axes[0].axvline(best_angle_deg, color='red', ls='--', label=f'Max variance at {best_angle_deg:.0f}°')
axes[0].set_xlabel('Projection angle (degrees)'); axes[0].set_ylabel('Variance of projections')
axes[0].set_title('Variance vs Projection Direction'); axes[0].legend(); axes[0].grid(alpha=0.3)

# Right: scatter + best direction
axes[1].scatter(X[:,0], X[:,1], alpha=0.4, s=15, color='gray')
scale = 2.5
axes[1].quiver(0, 0, best_direction[0]*scale, best_direction[1]*scale, 
               color='red', scale_units='xy', scale=1, width=0.03, label='Max variance direction')
axes[1].set_aspect('equal'); axes[1].grid(alpha=0.3)
axes[1].set_title('Data + Direction of Maximum Variance'); axes[1].legend()

plt.tight_layout(); plt.show()
# KEY INSIGHT: PCA finds this direction automatically using eigenvalues, not brute-force!
````
