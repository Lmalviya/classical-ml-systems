# Math Prerequisites → Linear Algebra | Chapter 2: Eigenvalues & Eigenvectors

> **Why this chapter exists:** PCA *is* eigendecomposition. The principal components are the eigenvectors of the covariance matrix, and the amount of variance they capture equals the corresponding eigenvalues. There is no shortcut — you cannot understand PCA deeply without understanding eigenvalues and eigenvectors from the ground up. This chapter builds from definition through full derivation to the Spectral Theorem.

---

## 1. Definition: What Are Eigenvectors and Eigenvalues?

For a square matrix $A \in \mathbb{R}^{n \times n}$, a non-zero vector $\vec{v} \in \mathbb{R}^n$ is an **eigenvector** of $A$ if multiplying it by $A$ only **scales** it — does not change its direction:

$$\boxed{A \vec{v} = \lambda \vec{v}}$$

The scalar $\lambda \in \mathbb{R}$ (or $\mathbb{C}$) is the **eigenvalue** corresponding to $\vec{v}$.

**In plain language:** Eigenvectors are the special directions in space that the matrix $A$ does not rotate — only stretches ($\lambda > 1$), shrinks ($0 < \lambda < 1$), preserves ($\lambda = 1$), flips ($\lambda < 0$), or collapses to zero ($\lambda = 0$).

### **Concrete 2D Example**
Let $A = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}$.

Try $\vec{v} = [1, 0]^T$: $A\vec{v} = [3, 0]^T = 3 \cdot [1,0]^T$. ✅ Eigenvector with $\lambda = 3$.

Try $\vec{v} = [0, 1]^T$: $A\vec{v} = [1, 2]^T \neq \lambda [0,1]^T$ for any scalar $\lambda$. ❌ Not an eigenvector.

### **The Eigenspace**
If $\vec{v}$ is an eigenvector, then so is any scalar multiple $c\vec{v}$:
$$A(c\vec{v}) = c(A\vec{v}) = c\lambda\vec{v} = \lambda(c\vec{v})$$

So eigenvectors are defined by **direction**, not magnitude. We conventionally normalize them to unit vectors.

---

## 2. Geometric Interpretation

Every matrix $A$ has certain **preferred directions** (eigenvectors) along which its action is pure scaling. All other directions get rotated.

### **Eigenvalue Sign and Magnitude**
| $\lambda$ | Effect on eigenvector $\vec{v}$ |
| :---: | :--- |
| $\lambda > 1$ | Stretches $\vec{v}$ away from origin |
| $\lambda = 1$ | $\vec{v}$ unchanged |
| $0 < \lambda < 1$ | Shrinks $\vec{v}$ toward origin |
| $\lambda = 0$ | Collapses $\vec{v}$ to zero (loses information) |
| $\lambda < 0$ | Flips direction AND scales |

### **Connection to PCA**
The covariance matrix $\Sigma$ has eigenvectors that point in the directions of maximum variance (these become the PCs), and eigenvalues that equal the variance along each direction:

$$\Sigma \vec{w}_k = \lambda_k \vec{w}_k$$

- $\vec{w}_k$: $k$-th principal component direction (unit vector)
- $\lambda_k$: variance of the data projected onto $\vec{w}_k$

Sorting by $\lambda_1 \geq \lambda_2 \geq \cdots \geq \lambda_d \geq 0$ gives PCs in order of importance.

---

## 3. Finding Eigenvalues: The Characteristic Polynomial

Start from $A\vec{v} = \lambda\vec{v}$. Rearrange:
$$(A - \lambda I)\vec{v} = \vec{0}$$

For this to have a **non-zero** solution $\vec{v}$, the matrix $(A - \lambda I)$ must be **singular** (not invertible). A matrix is singular iff its determinant is zero:

$$\boxed{\det(A - \lambda I) = 0}$$

This is the **characteristic equation**. Expanding the determinant gives a degree-$n$ polynomial in $\lambda$ — the **characteristic polynomial** $p(\lambda)$.

### **Full 2×2 Example**
Find eigenvalues of $A = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}$:

$$\det(A - \lambda I) = \det\begin{bmatrix} 4-\lambda & 2 \\ 1 & 3-\lambda \end{bmatrix} = (4-\lambda)(3-\lambda) - 2 \cdot 1$$
$$= 12 - 7\lambda + \lambda^2 - 2 = \lambda^2 - 7\lambda + 10 = 0$$
$$\Rightarrow (\lambda - 5)(\lambda - 2) = 0 \Rightarrow \lambda_1 = 5, \quad \lambda_2 = 2$$

**Check:** $\text{tr}(A) = 4 + 3 = 7 = \lambda_1 + \lambda_2 = 5 + 2$ ✅

$\det(A) = 12 - 2 = 10 = \lambda_1 \cdot \lambda_2 = 5 \times 2$ ✅

---

## 4. Finding Eigenvectors: The Null Space

For each eigenvalue $\lambda_k$, find $\vec{v}$ by solving $(A - \lambda_k I)\vec{v} = \vec{0}$ (the null space of $(A - \lambda_k I)$):

### **Continuing the 2×2 Example**
**For $\lambda_1 = 5$:**
$$A - 5I = \begin{bmatrix} 4-5 & 2 \\ 1 & 3-5 \end{bmatrix} = \begin{bmatrix} -1 & 2 \\ 1 & -2 \end{bmatrix}$$

Row reduce: both rows say $-v_1 + 2v_2 = 0$, so $v_1 = 2v_2$.

Let $v_2 = 1$: $\vec{v}_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}$. Normalized: $\hat{v}_1 = \frac{1}{\sqrt{5}}\begin{bmatrix} 2 \\ 1 \end{bmatrix}$

**Verify:** $A\vec{v}_1 = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}\begin{bmatrix} 2 \\ 1 \end{bmatrix} = \begin{bmatrix} 10 \\ 5 \end{bmatrix} = 5\begin{bmatrix} 2 \\ 1 \end{bmatrix}$ ✅

**For $\lambda_2 = 2$:**
$$A - 2I = \begin{bmatrix} 2 & 2 \\ 1 & 1 \end{bmatrix}$$

Row reduce: $v_1 + v_2 = 0$, so $v_1 = -v_2$.

Let $v_2 = 1$: $\vec{v}_2 = \begin{bmatrix} -1 \\ 1 \end{bmatrix}$. Normalized: $\hat{v}_2 = \frac{1}{\sqrt{2}}\begin{bmatrix} -1 \\ 1 \end{bmatrix}$

**Check orthogonality** (for symmetric matrices, eigenvectors of distinct eigenvalues are always orthogonal):
$$\vec{v}_1 \cdot \vec{v}_2 = (2)(-1) + (1)(1) = -1 \neq 0$$

> ⚠️ Note: matrix $A = \begin{bmatrix} 4 & 2 \\ 1 & 3 \end{bmatrix}$ is **not symmetric** ($A \neq A^T$), so eigenvectors need not be orthogonal. The covariance matrix IS symmetric, so its eigenvectors ARE always orthogonal.

---

## 5. The Spectral Theorem (For Symmetric Matrices)

The **Spectral Theorem** is the mathematical bedrock of PCA:

> **Theorem:** If $A \in \mathbb{R}^{n \times n}$ is symmetric ($A = A^T$), then:
> 1. All eigenvalues $\lambda_i$ are **real** (not complex).
> 2. Eigenvectors corresponding to **distinct** eigenvalues are **orthogonal**.
> 3. $A$ has a complete set of $n$ orthonormal eigenvectors — they form an **orthonormal basis** for $\mathbb{R}^n$.

### **Why This Guarantees PCA Works**
The covariance matrix $\Sigma$ is always symmetric. The Spectral Theorem guarantees:
- Real eigenvalues → variance in each direction is a real number ✅
- Orthogonal eigenvectors → PCs are independent (no redundancy) ✅
- Complete eigenbasis → PCs span the full feature space ✅

### **Symmetric Matrix Worked Example**
Find eigenvectors of $\Sigma = \begin{bmatrix} 3 & 1 \\ 1 & 3 \end{bmatrix}$ (symmetric):

**Characteristic equation:**
$$\det\begin{bmatrix} 3-\lambda & 1 \\ 1 & 3-\lambda \end{bmatrix} = (3-\lambda)^2 - 1 = \lambda^2 - 6\lambda + 8 = 0$$
$$\Rightarrow \lambda_1 = 4, \quad \lambda_2 = 2$$

**Eigenvector for $\lambda_1 = 4$:**
$$(\Sigma - 4I)\vec{v} = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}\vec{v} = \vec{0} \Rightarrow v_1 = v_2 \Rightarrow \hat{v}_1 = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix}$$

**Eigenvector for $\lambda_2 = 2$:**
$$(\Sigma - 2I)\vec{v} = \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}\vec{v} = \vec{0} \Rightarrow v_1 = -v_2 \Rightarrow \hat{v}_2 = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

**Orthogonality check:** $\hat{v}_1 \cdot \hat{v}_2 = \frac{1}{2}(1 \cdot 1 + 1 \cdot (-1)) = 0$ ✅ (Spectral Theorem confirmed)

**Interpretation:** This covariance matrix has maximum variance along the $[1,1]$ diagonal direction (variance = 4), and minimum variance along the $[1,-1]$ anti-diagonal (variance = 2). PCA would choose PC1 = $[1,1]/\sqrt{2}$.

---

## 6. Eigendecomposition: Factoring a Matrix via Its Eigenvectors

### **General Eigendecomposition**
If $A \in \mathbb{R}^{n \times n}$ has $n$ linearly independent eigenvectors $\vec{v}_1, ..., \vec{v}_n$ with eigenvalues $\lambda_1, ..., \lambda_n$:

Stack the equations $A\vec{v}_i = \lambda_i \vec{v}_i$ into one matrix equation:
$$A [\vec{v}_1 | \vec{v}_2 | \cdots | \vec{v}_n] = [\vec{v}_1 | \vec{v}_2 | \cdots | \vec{v}_n] \begin{bmatrix} \lambda_1 & & \\ & \ddots & \\ & & \lambda_n \end{bmatrix}$$
$$AQ = Q\Lambda \Rightarrow \boxed{A = Q\Lambda Q^{-1}}$$

### **For Symmetric Matrices (Spectral Decomposition)**
When $A = A^T$, the eigenvectors form an **orthonormal** matrix $Q$ (so $Q^{-1} = Q^T$):

$$\boxed{A = Q\Lambda Q^T = \sum_{i=1}^n \lambda_i \vec{q}_i \vec{q}_i^T}$$

This is the **Spectral Decomposition** — $A$ is a sum of rank-1 matrices, each scaled by its eigenvalue.

For the covariance matrix: $\Sigma = Q\Lambda Q^T$ where:
- $Q = [\vec{w}_1 | \vec{w}_2 | \cdots | \vec{w}_d]$ — columns are PCs (orthonormal)
- $\Lambda = \text{diag}(\lambda_1, ..., \lambda_d)$ — diagonal matrix of variances

### **The Rank-1 Outer Product Expansion**
$$\Sigma = \lambda_1 \vec{w}_1 \vec{w}_1^T + \lambda_2 \vec{w}_2 \vec{w}_2^T + \cdots + \lambda_d \vec{w}_d \vec{w}_d^T$$

Each term $\lambda_k \vec{w}_k \vec{w}_k^T$ represents the **contribution of PC $k$ to the total covariance structure**. Dropping the last $d-k$ terms gives the best rank-$k$ approximation to $\Sigma$ — the foundation for choosing how many PCs to keep.

### **Reconstructing Data**
Using eigendecomposition, reconstruction of $X$ from top-$k$ PCs:
$$\hat{X} = X Q_k Q_k^T$$

where $Q_k = [\vec{w}_1 | \cdots | \vec{w}_k]$ contains only the top $k$ eigenvectors.

---

## 7. Full Reference: Key Facts

| Concept | Formula / Fact |
| :--- | :--- |
| Eigenvalue equation | $A\vec{v} = \lambda\vec{v}$ |
| Characteristic equation | $\det(A - \lambda I) = 0$ |
| Sum of eigenvalues | $\sum \lambda_i = \text{tr}(A)$ |
| Product of eigenvalues | $\prod \lambda_i = \det(A)$ |
| Symmetric matrix | Real eigenvalues + orthogonal eigenvectors |
| PSD matrix | All $\lambda_i \geq 0$ |
| Eigendecomposition (symmetric) | $A = Q\Lambda Q^T$ |
| Spectral expansion | $A = \sum_i \lambda_i \vec{q}_i \vec{q}_i^T$ |
| PCA connection | PC $k$ = eigenvector $\vec{w}_k$ of $\Sigma$; variance = $\lambda_k$ |

---

## 8. Implementation Playground

```python
# ─── CELL 1: Full Manual Eigendecomposition — 2×2 and 3×3 ────────────────────
import numpy as np

# ── 2x2 symmetric matrix (covariance-like) ───────────────────────────────────
Sigma = np.array([[3., 1.], [1., 3.]])
print("=== 2x2 Symmetric Matrix ===")
print(f"\nΣ =");
print(Sigma)

# numpy eigvalsh is preferred for symmetric: guarantees real, sorted eigenvalues
eigenvalues, eigenvectors = np.linalg.eigh(Sigma)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print(f"\nEigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors.round(4)}")

# Verify: A @ v = lambda * v
for i in range(len(eigenvalues)):
    lhs = Sigma @ eigenvectors[:, i]
    rhs = eigenvalues[i] * eigenvectors[:, i]
    print(f"  v{i+1}: Av = {lhs.round(4)}, λv = {rhs.round(4)},  match={np.allclose(lhs, rhs)}")

# Verify: A = Q @ Lambda @ Q^T
Q = eigenvectors
Lambda = np.diag(eigenvalues)
A_reconstructed = Q @ Lambda @ Q.T
print(f"\nReconstructed from QΛQᵀ equals Σ? {np.allclose(A_reconstructed, Sigma)}")

# Orthonormality check
print(f"Qᵀ @ Q = I? {np.allclose(Q.T @ Q, np.eye(2))}")
print(f"Trace check: sum(λ)={eigenvalues.sum():.2f} = tr(Σ)={np.trace(Sigma):.2f}")

# ── 3x3 ─────────────────────────────────────────────────────────────────────
print("\n=== 3x3 Covariance Matrix ===")
np.random.seed(1)
X = np.random.randn(100, 3)
Sigma3 = X.T @ X / 99
eigenvalues3, eigenvectors3 = np.linalg.eigh(Sigma3)
eigenvalues3 = eigenvalues3[::-1]
eigenvectors3 = eigenvectors3[:, ::-1]
print(f"Eigenvalues: {eigenvalues3.round(4)}")
print(f"Explained variance: {(eigenvalues3/eigenvalues3.sum()*100).round(1)}%")
```

```python
# ─── CELL 2: Visualize Eigenvectors on 2D Data ────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
N = 400
# Correlated 2D data: covariance Sigma = [[3,1],[1,3]] effectively
mean = np.array([0., 0.])
cov_true = np.array([[3., 1.5], [1.5, 2.]])
X = np.random.multivariate_normal(mean, cov_true, N)

# Compute sample covariance and its eigenvectors
cov_sample = np.cov(X.T)  # 2x2
eigenvalues, eigenvectors = np.linalg.eigh(cov_sample)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Data + eigenvectors
ax = axes[0]
ax.scatter(X[:,0], X[:,1], s=8, alpha=0.3, color='gray')
colors = ['red', 'blue']
labels = ['PC1 (λ=%.2f)' % eigenvalues[0], 'PC2 (λ=%.2f)' % eigenvalues[1]]
for i in range(2):
    ev = eigenvectors[:, i]
    scale = np.sqrt(eigenvalues[i]) * 2.5  # scale arrow by std dev
    ax.quiver(0, 0, ev[0]*scale, ev[1]*scale,
              color=colors[i], scale_units='xy', scale=1, width=0.025, label=labels[i])
    ax.quiver(0, 0, -ev[0]*scale, -ev[1]*scale,
              color=colors[i], scale_units='xy', scale=1, width=0.025, alpha=0.4)
ax.set_aspect('equal'); ax.grid(alpha=0.3); ax.legend(fontsize=11)
ax.set_title('2D Data + Eigenvectors of Covariance Matrix', fontsize=12)

# Right: Rank-1 spectral expansion — show each PC's contribution separately
ax2 = axes[1]
x_grid = np.linspace(-4, 4, 50)
y_grid = np.linspace(-4, 4, 50)
XX, YY = np.meshgrid(x_grid, y_grid)
Z_total = np.zeros_like(XX)
for i in range(2):
    v = eigenvectors[:, i]
    rank1 = eigenvalues[i] * np.outer(v, v)
    for r in range(50):
        for c in range(50):
            z = np.array([XX[r,c], YY[r,c]])
            Z_total[r,c] += eigenvalues[i] * (z @ v)**2

ax2.contourf(XX, YY, Z_total, levels=20, cmap='RdYlBu_r', alpha=0.7)
ax2.scatter(X[:,0], X[:,1], s=8, alpha=0.2, color='black')
for i in range(2):
    ev = eigenvectors[:, i]
    scale = np.sqrt(eigenvalues[i]) * 2.5
    ax2.quiver(0, 0, ev[0]*scale, ev[1]*scale,
               color=colors[i], scale_units='xy', scale=1, width=0.025, label=labels[i])
ax2.set_aspect('equal'); ax2.grid(alpha=0.3); ax2.legend(fontsize=11)
ax2.set_title('Spectral Expansion: λ₁v₁v₁ᵀ + λ₂v₂v₂ᵀ', fontsize=12)

plt.tight_layout(); plt.show()
print(f"Eigenvalues: {eigenvalues.round(3)}")
print(f"EVR: PC1={eigenvalues[0]/eigenvalues.sum()*100:.1f}%, PC2={eigenvalues[1]/eigenvalues.sum()*100:.1f}%")
```

```python
# ─── CELL 3: Spectral Expansion — Rank-k Reconstruction of Image ──────────────
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_images

# Use sklearn's sample image or synthetic
try:
    img = load_sample_images().images[0][:,:,0].astype(float)  # grayscale
except:
    # Fallback: synthetic image
    np.random.seed(0)
    coords = np.mgrid[:100, :100].reshape(2,-1).T.astype(float)
    img = np.sin(coords[:,0]/5) * np.cos(coords[:,1]/5)
    img = img.reshape(100,100)

# Crop and normalize
M = img[:200, :200] if img.shape[0] >= 200 else img
M = M - M.mean()  # center

# Full SVD (related to eigendecomposition: eigenvalues of M^T M = sigma^2)
# Reconstruct with top-k eigenvalues/vectors
U, S, Vt = np.linalg.svd(M, full_matrices=False)

ks = [1, 5, 20, 50, 100]
fig, axes = plt.subplots(1, len(ks)+1, figsize=(4*(len(ks)+1), 4))

axes[0].imshow(M, cmap='gray'); axes[0].set_title('Original')
axes[0].axis('off')

for i, k in enumerate(ks):
    # Rank-k approximation using spectral expansion
    M_k = (U[:, :k] * S[:k]) @ Vt[:k, :]  # = sum_{i=1}^k sigma_i * u_i * v_i^T
    error = np.linalg.norm(M - M_k, 'fro')**2 / np.linalg.norm(M, 'fro')**2
    explained = (S[:k]**2).sum() / (S**2).sum() * 100
    axes[i+1].imshow(M_k, cmap='gray')
    axes[i+1].set_title(f'k={k}\n{explained:.1f}% variance')
    axes[i+1].axis('off')

plt.suptitle('Eigenvalue/SVD Rank-k Reconstruction\n(Spectral Expansion: M ≈ Σ σᵢ uᵢ vᵢᵀ)', fontsize=12)
plt.tight_layout(); plt.show()
print("\nKey: This is exactly what PCA does to DATA — keeps k largest eigenvalue directions.")
```
