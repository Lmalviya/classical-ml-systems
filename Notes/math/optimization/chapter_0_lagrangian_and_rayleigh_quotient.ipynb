# Math Prerequisites → Optimization | Chapter 0: Lagrangian & Rayleigh Quotient

> **Why this chapter exists:** The PCA optimization problem — "find the direction of maximum variance under a unit-norm constraint" — is a constrained optimization problem. This chapter derives its solution from first principles using the **Lagrangian method** and the **Rayleigh Quotient**. After this chapter, the PCA derivation in Chapter 0 of the PCA series will feel completely natural, not pulled from thin air.

---

## 1. Unconstrained Optimization — Quick Recap

For an unconstrained problem $\min_x f(x)$:
- **Necessary condition (first-order):** $f'(x^*) = 0$ at any local minimum or maximum.
- **Sufficient condition (second-order):** $f''(x^*) > 0$ at a minimum; $f''(x^*) < 0$ at a maximum.

In multiple dimensions: $\nabla f(\vec{x}^*) = \vec{0}$ (gradient = zero vector).

PCA is **not unconstrained** — we want to maximize variance but the direction must be a unit vector. Removing the constraint would let the solution blow up to infinity. We need constrained optimization.

---

## 2. Constrained Optimization: The Setting

A **constrained optimization problem** with equality constraints:
$$\min_{\vec{x}} f(\vec{x}) \quad \text{subject to} \quad g_i(\vec{x}) = 0, \quad i = 1, ..., m$$

For PCA:
$$\max_{\vec{w}} \vec{w}^T \Sigma \vec{w} \quad \text{subject to} \quad \vec{w}^T \vec{w} = 1$$

- **Objective:** $f(\vec{w}) = \vec{w}^T \Sigma \vec{w}$ (variance of projected data)
- **Constraint:** $g(\vec{w}) = \vec{w}^T \vec{w} - 1 = 0$ (unit length)

We cannot just take $\nabla f = 0$ — that gives the trivial solution $\vec{w} = \vec{0}$. We need the **Lagrangian method**.

---

## 3. The Lagrangian Method

### **The Key Idea**
At a constrained optimum, the gradient of the objective must be **parallel** to the gradient of the constraint. Otherwise we could move along the constraint surface and increase $f$. Mathematically:

$$\nabla f(\vec{x}^*) = \lambda \nabla g(\vec{x}^*)$$

where $\lambda$ is called the **Lagrange multiplier** — an unknown scalar we solve for.

### **The Lagrangian Function**
We combine the objective and constraint into a single function:
$$\mathcal{L}(\vec{x}, \lambda) = f(\vec{x}) - \lambda \cdot g(\vec{x})$$

The optimum satisfies:
$$\frac{\partial \mathcal{L}}{\partial \vec{x}} = \nabla f - \lambda \nabla g = \vec{0}$$
$$\frac{\partial \mathcal{L}}{\partial \lambda} = -g(\vec{x}) = 0 \quad \Rightarrow \quad g(\vec{x}) = 0$$

The second equation just restores the original constraint. The first gives the key stationarity condition.

---

## 4. ✍️ Deriving PCA from First Principles: Full Proof

**Problem:** Find unit vector $\vec{w}$ that maximizes the variance of data projected onto $\vec{w}$.

**Setup:** Given centered data $X \in \mathbb{R}^{N \times d}$, the variance of projections $\{\vec{x}_i \cdot \vec{w}\}$ is:
$$\text{Var}(\vec{w}) = \frac{1}{N-1} \sum_{i=1}^N (\vec{x}_i \cdot \vec{w})^2 = \frac{1}{N-1} \|X\vec{w}\|^2 = \vec{w}^T \underbrace{\left(\frac{X^T X}{N-1}\right)}_{\Sigma} \vec{w} = \vec{w}^T \Sigma \vec{w}$$

**Optimization problem:**
$$\max_{\vec{w}} \vec{w}^T \Sigma \vec{w} \quad \text{s.t.} \quad \vec{w}^T \vec{w} = 1$$

**Step 1: Form the Lagrangian:**
$$\mathcal{L}(\vec{w}, \lambda) = \vec{w}^T \Sigma \vec{w} - \lambda(\vec{w}^T \vec{w} - 1)$$

**Step 2: Take the gradient with respect to $\vec{w}$ and set to zero:**
$$\frac{\partial \mathcal{L}}{\partial \vec{w}} = 2\Sigma \vec{w} - 2\lambda \vec{w} = \vec{0}$$
$$\boxed{\Sigma \vec{w} = \lambda \vec{w}}$$

**This is the eigenvalue equation!** $\vec{w}$ must be an eigenvector of $\Sigma$ and $\lambda$ is the corresponding eigenvalue.

**Step 3: Identify which eigenvector maximizes the objective:**

Substitute back: the variance achieved by any eigenvector $\vec{w}_k$:
$$\vec{w}_k^T \Sigma \vec{w}_k = \vec{w}_k^T (\lambda_k \vec{w}_k) = \lambda_k \underbrace{\vec{w}_k^T \vec{w}_k}_{=1} = \lambda_k$$

The variance equals the eigenvalue! So to **maximize variance**, choose the eigenvector with the **largest eigenvalue** $\lambda_1$.

**Step 4: Subsequent PCs via deflation**

For PC2: repeat with the constraint $\vec{w}_2 \perp \vec{w}_1$ (add constraint $\vec{w}_1^T \vec{w}_2 = 0$).
The Lagrangian becomes:
$$\mathcal{L} = \vec{w}_2^T \Sigma \vec{w}_2 - \lambda(\vec{w}_2^T \vec{w}_2 - 1) - \phi (\vec{w}_1^T \vec{w}_2)$$

The solution: $\vec{w}_2$ = eigenvector with second largest eigenvalue $\lambda_2$.

**By induction: PC $k$ = eigenvector with $k$-th largest eigenvalue. Variance captured = $\lambda_k$.**

---

## 5. The Rayleigh Quotient: A Unified View

For a symmetric PSD matrix $A$, the **Rayleigh quotient** of a vector $\vec{x}$ is:
$$R_A(\vec{x}) = \frac{\vec{x}^T A \vec{x}}{\vec{x}^T \vec{x}}$$

**Key theorem:**
$$\lambda_{\min} \leq R_A(\vec{x}) \leq \lambda_{\max} \quad \forall \vec{x} \neq \vec{0}$$

The maximum is achieved at $\vec{x} = \vec{q}_1$ (eigenvector of largest eigenvalue), and the minimum at $\vec{x} = \vec{q}_n$ (eigenvector of smallest eigenvalue).

**For unit vectors** ($\vec{x}^T \vec{x} = 1$): $R_A(\vec{x}) = \vec{x}^T A \vec{x}$.

**So the PCA problem is literally:** find $\vec{w}$ that maximizes the Rayleigh quotient of $\Sigma$. Answer: first eigenvector of $\Sigma$.

### **The Min-Max Theorem (Courant-Fischer)**
The $k$-th largest eigenvalue of $A$ is:
$$\lambda_k = \max_{\vec{x} \perp \vec{q}_1,...,\vec{q}_{k-1}} \frac{\vec{x}^T A \vec{x}}{\vec{x}^T \vec{x}}$$

This is the exact greedy algorithm PCA uses: find the maximum variance direction, remove it, find the next maximum, and so on.

---

## 6. Summary: The PCA Proof in 4 Lines

| Step | Action | Result |
| :---: | :--- | :--- |
| 1 | Write the variance of projections | $\text{Var} = \vec{w}^T \Sigma \vec{w}$ |
| 2 | Form Lagrangian with unit-norm constraint | $\mathcal{L} = \vec{w}^T \Sigma \vec{w} - \lambda(\|\vec{w}\|^2 - 1)$ |
| 3 | Set gradient = 0 | $\Sigma \vec{w} = \lambda \vec{w}$ → eigenvalue equation |
| 4 | Variance = $\lambda$; maximize → choose largest | $\vec{w}_1 = $ top eigenvector of $\Sigma$ |

---

```python
# ─── CELL 1: Rayleigh Quotient — Visualize the Landscape ─────────────────────
import numpy as np
import matplotlib.pyplot as plt

# A 2x2 covariance matrix
Sigma = np.array([[4., 1.5], [1.5, 2.]])
eigenvalues, eigenvectors = np.linalg.eigh(Sigma)
eigenvalues = eigenvalues[::-1]; eigenvectors = eigenvectors[:, ::-1]

# Rayleigh quotient for all unit directions on the unit circle
angles = np.linspace(0, 2*np.pi, 360)
rayleigh = np.array([
    np.array([np.cos(a), np.sin(a)]) @ Sigma @ np.array([np.cos(a), np.sin(a)])
    for a in angles
])

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: Rayleigh quotient as function of angle
axes[0].plot(np.degrees(angles), rayleigh, color='steelblue', lw=2.5)
axes[0].axhline(eigenvalues[0], color='red', ls='--', lw=1.5, label=f'λ_max = {eigenvalues[0]:.2f}')
axes[0].axhline(eigenvalues[1], color='blue', ls='--', lw=1.5, label=f'λ_min = {eigenvalues[1]:.2f}')
max_angle = np.degrees(np.arctan2(eigenvectors[1,0], eigenvectors[0,0]))
axes[0].axvline(max_angle % 360, color='red', alpha=0.4, ls=':', label=f'PC1 direction ({max_angle%360:.1f}°)')
axes[0].set_xlabel('Direction angle (degrees)', fontsize=11)
axes[0].set_ylabel('Rayleigh Quotient wᵀΣw', fontsize=11)
axes[0].set_title('Rayleigh Quotient vs Direction\n(PCA finds argmax = top eigenvector)', fontsize=11)
axes[0].legend(); axes[0].grid(alpha=0.3)

# Right: Geometric view — the axes-stretching
theta = np.linspace(0, 2*np.pi, 200)
unit_circle = np.array([np.cos(theta), np.sin(theta)])
# Σ maps unit circle to an ellipse
ellipse = Sigma @ unit_circle

axes[1].plot(unit_circle[0], unit_circle[1], 'b--', lw=1.5, alpha=0.5, label='Unit circle (input)')
axes[1].plot(ellipse[0], ellipse[1], 'orange', lw=2, label='Σ × unit circle (output ellipse)')
for i, color in enumerate(['red','blue']):
    ev = eigenvectors[:, i]
    lam = eigenvalues[i]
    axes[1].quiver(0, 0, ev[0]*lam, ev[1]*lam,
                   color=color, scale_units='xy', scale=1, width=0.025,
                   label=f'Eigenvector {i+1} (λ={lam:.2f})')
axes[1].set_aspect('equal'); axes[1].grid(alpha=0.3)
axes[1].set_title('Geometric View: Σ Stretches Unit Circle to Ellipse\nEigenvectors = Ellipse Axes', fontsize=11)
axes[1].legend(fontsize=9)

plt.tight_layout(); plt.show()
print(f"Maximum Rayleigh quotient: {rayleigh.max():.4f}  =  λ₁: {eigenvalues[0]:.4f} ✓")
print(f"Minimum Rayleigh quotient: {rayleigh.min():.4f}  =  λ₂: {eigenvalues[1]:.4f} ✓")
```

```python
# ─── CELL 2: Verifying the Lagrangian Derivation — Gradient at Solution ────────
import numpy as np

# Covariance matrix
Sigma = np.array([[4., 1.5], [1.5, 2.]])
eigenvalues, eigenvectors = np.linalg.eigh(Sigma)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]; eigenvectors = eigenvectors[:, idx]

w_opt = eigenvectors[:, 0]  # top eigenvector
lambda_opt = eigenvalues[0]

print("=== Verifying PCA Derivation: Σw = λw ===")
print(f"Top eigenvector w₁ = {w_opt.round(4)}")
print(f"Corresponding eigenvalue λ₁ = {lambda_opt:.4f}")

lhs = Sigma @ w_opt          # Σw
rhs = lambda_opt * w_opt     # λw
print(f"\nΣw  = {lhs.round(4)}")
print(f"λ·w = {rhs.round(4)}")
print(f"Σw = λw? {np.allclose(lhs, rhs)}")

# Verify: variance achieved = eigenvalue
variance = w_opt @ Sigma @ w_opt
print(f"\nVariance wᵀΣw = {variance:.4f}  =  λ₁ = {lambda_opt:.4f}? {np.allclose(variance, lambda_opt)}")

# Verify: Lagrangian gradient = 0 at solution
# ∂L/∂w = 2Σw - 2λw = 2(Σ - λI)w = 0
gradient_L = 2 * (Sigma - lambda_opt * np.eye(2)) @ w_opt
print(f"\nLagrangian gradient at solution: {gradient_L.round(10)}")
print(f"Gradient = 0 (stationarity)? {np.allclose(gradient_L, 0)}")

# Show all eigenvectors satisfy the eigenvalue equation
print("\n=== All Eigenvectors Satisfy Σwₖ = λₖwₖ ===")
for k in range(len(eigenvalues)):
    wk = eigenvectors[:, k]
    print(f"  k={k+1}: ||Σwₖ - λₖwₖ|| = {np.linalg.norm(Sigma @ wk - eigenvalues[k]*wk):.2e}")
```
