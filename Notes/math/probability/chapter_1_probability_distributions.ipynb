# Math Prerequisites → Probability | Chapter 1: Probability Distributions

> **Why this chapter exists:** Every Bayesian model requires choosing a **likelihood** (what distribution generated the data?) and a **prior** (what distribution represents our beliefs about parameters?). This chapter builds fluency with the key distributions used throughout the Bayesian ML series — Bernoulli, Binomial, Categorical, Multinomial (likelihoods), and Beta, Dirichlet, Gaussian (priors and posteriors).

---

## 1. Discrete Distributions

### **Bernoulli Distribution**
Models a single binary outcome (coin flip, spam/ham, click/no-click).

$$X \sim \text{Bernoulli}(\theta), \quad x \in \{0, 1\}$$
$$P(X = x) = \theta^x (1-\theta)^{1-x}$$

- **Mean:** $E[X] = \theta$
- **Variance:** $\text{Var}(X) = \theta(1-\theta)$ (maximized at $\theta = 0.5$)
- **Parameter:** $\theta \in [0,1]$ (probability of success)

---

### **Binomial Distribution**
Number of successes in $n$ independent Bernoulli trials.

$$X \sim \text{Binomial}(n, \theta), \quad x \in \{0, 1, ..., n\}$$
$$P(X = k) = \binom{n}{k} \theta^k (1-\theta)^{n-k}$$

- **Mean:** $E[X] = n\theta$
- **Variance:** $\text{Var}(X) = n\theta(1-\theta)$
- **Bernoulli(θ) = Binomial(1, θ)**

---

### **Categorical Distribution**
Generalizes Bernoulli to $K$ outcomes. Models: which word in a vocabulary, which class label.

$$X \sim \text{Categorical}(\boldsymbol{\theta}), \quad x \in \{1, 2, ..., K\}$$
$$P(X = k) = \theta_k, \quad \sum_{k=1}^K \theta_k = 1$$

Parameters: $\boldsymbol{\theta} = (\theta_1, ..., \theta_K)$ — probability vector (simplex).

---

### **Multinomial Distribution**
Generalizes Binomial to $K$ outcomes: counts $(n_1, ..., n_K)$ from $n$ trials.

$$P(n_1, ..., n_K | n, \boldsymbol{\theta}) = \frac{n!}{n_1! \cdots n_K!} \prod_{k=1}^K \theta_k^{n_k}, \quad \sum_k n_k = n$$

**Used in:** Multinomial Naive Bayes for text — word count vectors in documents.

---

### **Poisson Distribution**
Number of events in a fixed interval given a constant rate $\lambda$.

$$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad k = 0, 1, 2, ...$$

**Mean = Variance = $\lambda$** (unique property). Used for: count data (clicks, typos, arrivals).

---

## 2. Continuous Distributions

### **Gaussian (Normal) Distribution**
$$X \sim \mathcal{N}(\mu, \sigma^2)$$
$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$$

- **Mean:** $\mu$, **Variance:** $\sigma^2$, **Mode:** $\mu$
- **68-95-99.7 rule:** $P(\mu \pm \sigma) = 68\%$, $P(\mu \pm 2\sigma) = 95\%$, $P(\mu \pm 3\sigma) = 99.7\%$
- **Stability:** Sum of independent Gaussians is Gaussian. $X+Y \sim \mathcal{N}(\mu_X+\mu_Y, \sigma_X^2+\sigma_Y^2)$
- **Used as:** Likelihood for continuous data, prior for Gaussian regression weights

---

### **Beta Distribution**
$$\theta \sim \text{Beta}(\alpha, \beta), \quad \theta \in [0,1]$$
$$p(\theta) = \frac{\theta^{\alpha-1}(1-\theta)^{\beta-1}}{B(\alpha, \beta)}, \quad B(\alpha,\beta) = \frac{\Gamma(\alpha)\Gamma(\beta)}{\Gamma(\alpha+\beta)}$$

- **Mean:** $\mu = \frac{\alpha}{\alpha+\beta}$ — ratio of "successes" to total count
- **Mode:** $\frac{\alpha-1}{\alpha+\beta-2}$ (for $\alpha, \beta > 1$)
- **Variance:** $\frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)}$
- **Key property:** Lives on $[0,1]$ → perfect prior for probabilities!
- **Conjugate prior** for the Bernoulli/Binomial likelihood (Chapter 3):
  - $\text{Beta}(\alpha, \beta)$ prior + $k$ successes out of $n$ → $\text{Beta}(\alpha+k, \beta+n-k)$ posterior

**Interpretation of hyperparameters:**
| Shape | Effect |
| :--- | :--- |
| $\alpha = \beta = 1$ | Uniform — no prior information |
| $\alpha = \beta > 1$ | Symmetric, peaked at 0.5 — believe fair coin |
| $\alpha > \beta$ | Skewed right — believe higher probability |
| $\alpha, \beta < 1$ | U-shaped — believe it's either 0 or 1 (bimodal) |

---

### **Dirichlet Distribution**
Multivariate generalization of Beta — prior over probability vectors (simplices).

$$\boldsymbol{\theta} \sim \text{Dir}(\boldsymbol{\alpha}), \quad \boldsymbol{\theta} \in \Delta^{K-1}$$
$$p(\boldsymbol{\theta}) = \frac{\Gamma(\sum_k \alpha_k)}{\prod_k \Gamma(\alpha_k)} \prod_{k=1}^K \theta_k^{\alpha_k - 1}$$

- **Mean:** $E[\theta_k] = \frac{\alpha_k}{\sum_j \alpha_j}$
- **Conjugate prior** for Categorical/Multinomial: $\text{Dir}(\boldsymbol{\alpha})$ + counts $\boldsymbol{n}$ → $\text{Dir}(\boldsymbol{\alpha} + \boldsymbol{n})$
- **Concentration parameter** $\sum_k \alpha_k$: large = concentrated near mean; small = spread across simplex
- **Used in:** Multinomial Naive Bayes smoothing, topic models (LDA)

---

### **Gamma Distribution**
$$X \sim \text{Gamma}(a, b), \quad x > 0$$
$$p(x) = \frac{b^a x^{a-1} e^{-bx}}{\Gamma(a)}$$

- **Mean:** $a/b$, **Variance:** $a/b^2$
- **Used as:** Conjugate prior for precision ($1/\sigma^2$) of a Gaussian — important in Bayesian inference
- Special case: $\text{Gamma}(1, b) = \text{Exponential}(b)$

---

## 3. The Exponential Family — A Unified View

Most useful distributions in ML belong to the **exponential family**:

$$p(x | \boldsymbol{\eta}) = h(x) \exp\left(\boldsymbol{\eta}^T \boldsymbol{T}(x) - A(\boldsymbol{\eta})\right)$$

where:
- $\boldsymbol{\eta}$ — **natural parameters** (canonical parameters)
- $\boldsymbol{T}(x)$ — **sufficient statistics** (capture all info about $\boldsymbol{\eta}$ from data)
- $A(\boldsymbol{\eta})$ — **log-partition function** (normalization, ensures $\int p = 1$)
- $h(x)$ — base measure

### **Why It Matters for ML**
| Property | Machine Learning Consequence |
| :--- | :--- |
| **Sufficient statistics** are finite-dimensional | MLE only needs $\boldsymbol{T}(x)$, not all raw data (e.g., just mean and variance for Gaussian) |
| **Always has conjugate priors** | Bayesian inference remains tractable for all exponential family likelihoods |
| **MLE = method of moments** | Set $E[T(x)] = $ sample mean of $T(x)$ |
| **Log-likelihood is concave** | MLE has a unique global maximum — efficient optimization |
| Generalized Linear Models (GLMs) | All GLMs (linear, logistic, Poisson regression) use exponential family likelihoods |

### **Quick Reference**
| Distribution | Natural param $\eta$ | Sufficient stat $T(x)$ |
| :--- | :--- | :--- |
| Bernoulli($\theta$) | $\log(\theta/(1-\theta))$ | $x$ |
| Gaussian($\mu$,$\sigma^2$) | $[\mu/\sigma^2, -1/(2\sigma^2)]$ | $[x, x^2]$ |
| Poisson($\lambda$) | $\log\lambda$ | $x$ |
| Categorical($\boldsymbol{\theta}$) | $\log\boldsymbol{\theta}$ | one-hot $x$ |

---

## 4. Choosing the Right Distribution

| Data Type | Likelihood Distribution | Example Use |
| :--- | :--- | :--- |
| Binary outcome | Bernoulli | Spam/ham, click/no-click |
| Count of successes in $n$ trials | Binomial | $k$ heads in $n$ flips |
| One of $K$ categories | Categorical | Class labels |
| Word count vector | Multinomial | Document in Naive Bayes |
| Count events per interval | Poisson | Emails per day |
| Continuous unbounded | Gaussian | Height, weight, regression noise |
| Continuous positive valued | Gamma / Log-Normal | Income, time-to-event |

| Parameter Type | Prior Distribution | Why |
| :--- | :--- | :--- |
| Probability $\theta \in [0,1]$ | Beta | Conjugate to Bernoulli |
| Probability vector $\boldsymbol{\theta} \in \Delta$ | Dirichlet | Conjugate to Multinomial |
| Mean of Gaussian (known var) | Gaussian | Conjugate to Gaussian |
| Precision of Gaussian | Gamma | Conjugate to Gaussian |
| Regression weights | Gaussian | → MAP = Ridge regression |

```python
# ─── CELL 1: Visualize All Key Distributions ──────────────────────────────────
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

fig, axes = plt.subplots(2, 4, figsize=(18, 9))
axes = axes.flatten()

# 1. Bernoulli
ax = axes[0]
for theta, color in [(0.3, 'steelblue'), (0.5, 'forestgreen'), (0.7, 'darkorange')]:
    ax.bar([0, 1], [1-theta, theta], width=0.2, alpha=0.7, color=color, label=f'θ={theta}')
ax.set_title('Bernoulli(θ)', fontsize=11); ax.set_xticks([0,1]); ax.set_xticklabels(['0','1'])
ax.legend(fontsize=8); ax.grid(alpha=0.3, axis='y')

# 2. Binomial
ax = axes[1]
for n, theta, color in [(10, 0.5, 'steelblue'), (20, 0.3, 'darkorange'), (30, 0.7, 'forestgreen')]:
    k = np.arange(0, n+1)
    ax.plot(k, stats.binom.pmf(k, n, theta), 'o-', lw=1.5, ms=4, color=color, label=f'n={n},θ={theta}')
ax.set_title('Binomial(n,θ)', fontsize=11); ax.legend(fontsize=8); ax.grid(alpha=0.3)

# 3. Poisson
ax = axes[2]
k = np.arange(0, 20)
for lam, color in [(1, 'steelblue'), (4, 'forestgreen'), (8, 'darkorange')]:
    ax.plot(k, stats.poisson.pmf(k, lam), 'o-', lw=1.5, ms=4, color=color, label=f'λ={lam}')
ax.set_title('Poisson(λ)', fontsize=11); ax.legend(fontsize=8); ax.grid(alpha=0.3)

# 4. Gaussian
ax = axes[3]
x = np.linspace(-5, 5, 300)
for mu, sigma, color in [(0, 1, 'steelblue'), (0, 2, 'darkorange'), (1, 0.5, 'forestgreen')]:
    ax.plot(x, stats.norm.pdf(x, mu, sigma), lw=2.5, color=color, label=f'μ={mu},σ={sigma}')
ax.set_title('Gaussian(μ,σ²)', fontsize=11); ax.legend(fontsize=8); ax.grid(alpha=0.3)

# 5. Beta
ax = axes[4]
x = np.linspace(0.001, 0.999, 300)
configs = [(1,1,'Uniform','gray'), (2,2,'Sym, flat','steelblue'),
           (5,2,'Right skew','darkorange'), (0.5,0.5,'U-shaped','forestgreen')]
for a, b, label, color in configs:
    ax.plot(x, stats.beta.pdf(x, a, b), lw=2.5, color=color, label=f'α={a},β={b} ({label})')
ax.set_ylim(0, 4); ax.set_title('Beta(α,β) — Prior for θ', fontsize=11)
ax.legend(fontsize=7); ax.grid(alpha=0.3)

# 6. Gamma
ax = axes[5]
x = np.linspace(0.001, 15, 300)
for a, b, color in [(1, 1, 'steelblue'), (2, 1, 'darkorange'), (5, 1, 'forestgreen'), (2, 0.5, 'purple')]:
    ax.plot(x, stats.gamma.pdf(x, a=a, scale=1/b), lw=2, color=color, label=f'a={a},b={b}')
ax.set_title('Gamma(a,b)', fontsize=11); ax.legend(fontsize=8); ax.grid(alpha=0.3)

# 7. Beta updating (prior → posterior after data)
ax = axes[6]
x = np.linspace(0.001, 0.999, 300)
# Start: uniform prior Beta(1,1); observe k=7 heads in n=10 flips
k, n = 7, 10
for alpha0, beta0, label, color in [(1, 1, 'Prior: Beta(1,1)', 'gray'),
                                      (1+k, 1+n-k, f'Posterior: Beta({1+k},{1+n-k})', 'steelblue'),
                                      (5, 5, 'Prior: Beta(5,5) (strong)', 'lightcoral')]:
    ax.plot(x, stats.beta.pdf(x, alpha0, beta0), lw=2.5, color=color, label=label)
# Also plot: strong prior posterior
ax.plot(x, stats.beta.pdf(x, 5+k, 5+n-k), lw=2.5, color='red', ls='--',
        label=f'Post from strong prior: Beta({5+k},{5+n-k})')
ax.axvline(k/n, color='black', ls=':', lw=1.5, label=f'MLE = {k/n:.1f}')
ax.set_title(f'Beta updating: k={k} heads in n={n}\nUniform vs strong prior', fontsize=10)
ax.legend(fontsize=7); ax.grid(alpha=0.3)

# 8. Dirichlet (shown as 2D simplex projection)
ax = axes[7]
np.random.seed(42)
configs_dir = [([1,1,1], 'Dir(1,1,1) — Uniform', 'steelblue'),
               ([5,5,5], 'Dir(5,5,5) — Concentrated', 'darkorange'),
               ([0.5,0.5,0.5], 'Dir(0.5,…) — Corners', 'forestgreen')]
for alpha, label, color in configs_dir:
    samples = np.random.dirichlet(alpha, 300)
    # Project to 2D using barycentric coordinates for visualization
    x2d = samples[:,0] + 0.5*samples[:,1]
    y2d = (np.sqrt(3)/2)*samples[:,1]
    ax.scatter(x2d, y2d, s=5, alpha=0.4, color=color, label=label)
ax.set_title('Dirichlet samples (K=3)\n(each point = probability vector)', fontsize=10)
ax.legend(fontsize=7); ax.set_xlim(0, 1.1); ax.set_ylim(0, 1); ax.grid(alpha=0.3)

plt.suptitle('Probability Distributions: The Building Blocks of Bayesian ML', fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()
```

```python
# ─── CELL 2: Beta Distribution — Prior Strength and Posterior Concentration ────
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(0)
true_theta = 0.65  # true coin bias
x = np.linspace(0.001, 0.999, 500)

# Different priors and their posteriors after N observations
priors = [(1, 1, 'Weak: Beta(1,1) — Uniform'), (3, 3, 'Moderate: Beta(3,3)'), (20, 20, 'Strong: Beta(20,20)')]
N_list = [5, 20, 100]

fig, axes = plt.subplots(len(priors), len(N_list), figsize=(14, 10))

for row, (a0, b0, prior_label) in enumerate(priors):
    for col, N in enumerate(N_list):
        # Sample N coin flips
        k = int(round(true_theta * N))  # observed heads
        # Posterior: Beta(a0+k, b0+N-k)
        a_post = a0 + k
        b_post = b0 + N - k

        ax = axes[row][col]
        ax.plot(x, stats.beta.pdf(x, a0, b0), 'k--', lw=1.5, alpha=0.5, label=f'Prior: Beta({a0},{b0})')
        ax.fill_between(x, stats.beta.pdf(x, a_post, b_post), alpha=0.35, color='steelblue')
        ax.plot(x, stats.beta.pdf(x, a_post, b_post), color='steelblue', lw=2.5,
                label=f'Posterior: Beta({a_post},{b_post})')
        ax.axvline(true_theta, color='red', ls='--', lw=1.5, label=f'True θ={true_theta}')
        ax.axvline(k/N, color='green', ls=':', lw=1.5, label=f'MLE={k/N:.2f}')
        # Posterior mean
        post_mean = a_post / (a_post + b_post)
        ax.axvline(post_mean, color='purple', ls='-', lw=1.5, alpha=0.6)

        ax.set_title(f'{prior_label}\nN={N}, k={k} heads', fontsize=8)
        ax.set_xlim(0, 1); ax.grid(alpha=0.3)
        if col == 0: ax.set_ylabel('Density', fontsize=9)
        if row == len(priors)-1: ax.set_xlabel('θ (probability of heads)', fontsize=9)
        if row == 0 and col == 0: ax.legend(fontsize=7)

plt.suptitle('Beta-Binomial Updating: Prior Strength vs Data Size\n'
             'Red=True θ, Green=MLE, Purple=Posterior mean, Blue region=Posterior',
             fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()

# Key observation
print("Key takeaways:")
print("  - Weak prior (1,1): posterior driven entirely by data — converges fast to MLE")
print("  - Strong prior (20,20): posterior pulled toward 0.5 — needs more data to 'override'")
print("  - As N increases, all priors give the same posterior (data dominates prior)")
print(f"  - Posterior mean = (α₀+k)/(α₀+β₀+N) = weighted avg of prior mean and MLE")
```
